```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
dataset=pd.read_csv("Iris.csv")
dataset.head(30)
```

```
dataset.info()
```

```
dataset.isna().sum()
dataset.info
```

```
#Let's perform EDA(Exploratory Data Analysis)
import seaborn as sns
sns.pairplot(dataset,hue="Species")
```

```
X_val=dataset[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y_val=dataset["Species"]
y_val
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_val,y_val,test_size=0.30,
                                               random_state=0)
```

```
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion="entropy",random_state=0)
classifier.fit(X_train,y_train)
```

```
y_pred=classifier.predict(X_test)
```

```
#Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
features=list(dataset.columns[2:])
features
```

```
dot_data = StringIO()
export_graphviz(classifier, out_file=dot_data, feature_names=features,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
